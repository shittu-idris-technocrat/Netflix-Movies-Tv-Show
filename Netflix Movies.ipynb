# 3🔹 Q3: Content added per year
    

```python
import matplotlib.pyplot as plt
```

```python
df["year_added"].value_counts().sort_index().plot(kind="bar", figsize=(10,5),title="Content Added per Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles Added")
plt.grid(True)
plt.show()
```

```python
df["country"].value_counts().head(10).sort_index().plot(kind="bar", figsize=(10, 5), title="Content Per Year")
plt.xlabel("Year")
plt.ylabel("Num of Titles")
plt.grid(True)
plt.show()
```

```python
#3 🔹 Q4: What are the most common countries for content?
```

```python
top_countries = df["country"].value_counts().head(10)
top_countries.plot(kind="bar", title="Top 10 Countries", figsize=(10, 15), color="coral")
plt.xlabel("country")
plt.ylabel("Number of Titles")
plt.grid(True)
plt.show()
```

```python
## 🔹 Q5: What’s the trend of Movies vs TV Shows over time?
```

```python
type_year = df.groupby(["year_added", "type"]).size().unstack()
type_year.plot(kind="line", figsize=(10,6), marker="o", title="Movies vs Tv shows")
plt.xlabel("Year")
plt.ylabel("Numbers of Titles")
plt.grid(True)
plt.show()
```

```python
top_directors=df[df["director"] != "unknown"]["director"].value_counts().head(10)
top_directors.plot(kind="barh", title="Top Directors  on Netflix", color="skyblue")
plt.xlabel("Numbers of Titles")
plt.gca().invert_yaxis()
plt.show()
```

# 📊 Netflix Movies & TV Shows Analysis

This project analyzes a dataset of Netflix titles to uncover trends and insights, using Python and Pandas.

## 📌 Key Insights
- Netflix has more movies than TV shows.
- Most content comes from the US, India, and UK.
- A significant spike in content occurred post-2016 and starting declining from 2020
- Most frequent genres include Dramas, Comedies, and Documentaries.
- Director with the most titles: [Raul Campos][Jan Suter].

- Python (Pandas, Matplotlib)
